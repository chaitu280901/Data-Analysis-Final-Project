```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
sns.set(style="white")
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
# Replace 'your_file.csv' with the actual path to your CSV file
file_path = r"D:\Data analysis project\autos.csv"

# Try reading the CSV file with a different encoding
try:
    df = pd.read_csv(file_path, encoding='utf-8')
except UnicodeDecodeError:
    # If reading with utf-8 fails, try other encodings
    try:
        df = pd.read_csv(file_path, encoding='ISO-8859-1')
    except UnicodeDecodeError:
        # If 'ISO-8859-1' also fails, try 'utf-16'
        df = pd.read_csv(file_path, encoding='utf-16')

# Continue with the rest of your data processing using the 'df' DataFrame
```

```python
df # Raw Data
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['nrOfPictures'],inplace=True,axis=1) ## All records in 'nrOfPictures' column are having 0s hence it is of no use so, I dropped them.
```

## Changing str to datetime

```python
df['dateCrawled']= pd.to_datetime(df['dateCrawled']) ## Changing str to datetime
```

```python
df['dateCreated']= pd.to_datetime(df['dateCreated']) # ........................
```

```python
df['lastSeen']= pd.to_datetime(df['lastSeen'])       # ........................
```

## Replacing and Updating

```python
df['notRepairedDamage'].replace({'ja':'Yes','nein':'No'},inplace=True)            ## Translated German to English
```

```python
df['offerType'].replace({'Angebot':'Offer','Gesuch':'Request'},inplace=True)
```

```python
df['vehicleType'].replace({'kleinwagen':'compact car','cabrio':'convertible'},inplace=True)                        
```

```python
df['seller'].replace({'privat':'private','gewerblich':'commercial'},inplace=True)
```

```python
df['gearbox'].replace({'manuell':'Manual','automatik':'Automatic'},inplace=True)
```

```python
df['fuelType'].replace('benzin','petrol',inplace=True)
```

```python
df['monthOfRegistration'].replace(0,1,inplace=True) ## Replacing 0 with 1 since 0 doesn't refer to any month.
```

## Removing Null Values

```python
df['model'].fillna('Other',inplace=True)
```

```python
df['gearbox'].fillna('Unspecified',inplace=True)
```

```python
df['vehicleType'].fillna('Other',inplace=True)
```

```python
df['fuelType'].fillna('Other',inplace=True)
```

```python
df['notRepairedDamage'].fillna('Other',inplace=True)
```

```python
df['monthOfRegistration'].replace({1:'January',2:'February',3:'March',4:'April',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'},inplace=True)
```

```python
def in_name(x):
    return re.sub(r'[_]',' ',x)
```

```python
df['name']=df['name'].apply(in_name) ## Removing '_' from name column.
```

```python
def in_model(x):
    return re.sub(r'[_]+?',' ',x)
```

```python
df['model']=df['model'].apply(in_model) ## Removing '_' from model column.
```

```python
df.isnull().sum()
```

## Removing Duplicates

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

## Changing the column order

```python
col_order=['name','seller','model','brand','offerType','price','abtest','vehicleType','gearbox','powerPS','fuelType','kilometer','notRepairedDamage','yearOfRegistration','monthOfRegistration','postalCode','dateCrawled','dateCreated','lastSeen']
```

```python
df=df[col_order]
```
